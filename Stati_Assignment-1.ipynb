1. Simulate exam scores for 500 students with mean = 70, std dev = 10. 
Tasks: Probability of scoring above 80, below 50, between 60â€“75. 
Visuals: Histogram with fitted  curve.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


mean = 70
std_dev = 10
total_students = 500

score = np.random.normal(mean, std_dev, total_students)


print(f"Probability of scoring above 80: {np.mean(score > 80):.2f}\n"
      f"Probability of scoring below 60: {np.mean(score < 60):.2f}\n"
      f"Probability of scoring between 65 and 75: {np.mean((score >= 65) & (score <= 75)):.2f}")



plt.figure(figsize=(10,6))
plt.hist(score, bins=30, density=True, color='lightcoral', edgecolor='black')
x = np.linspace(min(score), max(score), 100)
plt.plot(x, norm.pdf(x, mean, std_dev), 'r', lw=2)
plt.title('Distribution of Student Scores')
plt.xlabel('Scores')
plt.ylabel('Density')
plt.grid()
plt.show()
```

3. Customer First Purchase Day 
Simulate the day on which a new website visitor makes their first purchase (probability 
per day = 0.1). 
Find probability that a customer buys on Day 1, Day 3, etc. 
Visualize with PMF.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

p = 0.1
days = np.arange(1, 21)

pmf = geom.pmf(days, p)

for d in [1, 3, 5, 10]:
	print(f"Probability customer buys on Day {d}: {pmf[d-1]:.4f}")

plt.figure(figsize=(10,6))
plt.plot(days, pmf, marker='o')
plt.title('PMF: First Purchase Day')
plt.xlabel('Day')
plt.ylabel('Probability')
plt.grid()
plt.show()
```

4.Basketball Free Throw Attempts 
Probability of making a free throw = 0.6. 
Find the probability distribution of the number of attempts until the first basket. 
Discuss expectation and variance.

```python
p = 0.6

attempts = np.arange(1, 16)
pmf_attempts = geom.pmf(attempts, p)

for attempts, prob in enumerate(pmf_attempts,1):
    print(f"Probability of first basket on attempt {attempts}: {prob:.4f}")

expectation = geom.mean(p)
variance = geom.var(p)

print(f"Expected number of attempts until first basket: {expectation:.2f}")
print(f"Variance: {variance:.2f}")

plt.figure(figsize=(10,6))
plt.bar(np.arange(1, attempts+1), pmf_attempts, color='lightblue', edgecolor='black')
plt.title('PMF: Attempts Until First Basket (p=0.6)')
plt.xlabel('Number of Attempts')
plt.ylabel('Probability')
plt.grid()
plt.show()
```

5.Customer Arrival at a Coffee Shop 
Average arrivals = 5 customers/hour. 
Find probability of exactly 3, less than 2, more than 8 arrivals. 
Plot Poisson PMF and compare with Normal approximation.

```python
from scipy.stats import poisson, norm

avg_ar = 5


prob_3 = poisson.pmf(3, avg_ar)
prob_lt_2 = poisson.cdf(1, avg_ar)
prob_gt_8 = 1 - poisson.cdf(8, avg_ar)

print(f"Probability of exactly 3 arrivals: {prob_3:.4f}")
print(f"Probability of less than 2 arrivals: {prob_lt_2:.4f}")
print(f"Probability of more than 8 arrivals: {prob_gt_8:.4f}")


x = np.arange(0, 15)
poisson_pmf = poisson.pmf(x, avg_ar)
normal_approx = norm.pdf(x, loc=avg_ar, scale=np.sqrt(avg_ar))

plt.figure(figsize=(10,6))
plt.bar(x, poisson_pmf, color='lightgreen', edgecolor='black', alpha=0.7, label='Poisson PMF')
plt.plot(x, normal_approx, 'r--', lw=2, label='Normal Approximation')
plt.title('Customer Arrivals: Poisson vs Normal')
plt.xlabel('Number of Arrivals')
plt.ylabel('Probability')
plt.legend()
plt.grid()
plt.show()
```
